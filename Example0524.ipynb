```python
class Solution:#默認code
    def twoSum(self, nums: List[int], target: int) -> List[int]:#默認code
    dict = {}
    for i,n in enumerate(nums):
        if n in dict:
            return dict[n],i
        else:
            dict[target-n] = i
'''
    store = dict()
    for i in range(len(nums)):
        sec = target - nums[i]
        if sec not in store:
            store[nums[i]]=i
        else:
            return [store[sec],i]
''' 
```
